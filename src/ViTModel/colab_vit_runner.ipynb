```python
# Point this to your project root in Drive
# Example: /content/drive/MyDrive/Deep-learning-ALS
PROJECT_ROOT = '/content/drive/MyDrive/Deep-learning-ALS'

import os
import sys

SRC_DIR = os.path.join(PROJECT_ROOT, 'src')
if SRC_DIR not in sys.path:
    sys.path.insert(0, SRC_DIR)

print('Using src path:', SRC_DIR)
```

```python
# Paths for input CNN features and ViT outputs
FEATURES_DIR = os.path.join(SRC_DIR, 'cnn_features', 'cnn_multimodal', 'features')
ARTIFACTS_DIR = os.path.join(SRC_DIR, 'cnn_features', 'vit_multimodal')

print('FEATURES_DIR:', FEATURES_DIR)
print('ARTIFACTS_DIR:', ARTIFACTS_DIR)
```

```python
# Quick check: can we import the existing .py modules?
from ViTModel.model import SimpleMultiModalViT
from ViTModel.dataset import ALSFeatureDataset
from ViTModel.train_vit import train
from ViTModel.evaluate_vit import evaluate

print('Imports successful.')
```

```python
# Optional check: dataset summary before training
ds = ALSFeatureDataset(FEATURES_DIR)
print('Num samples:', len(ds))
print('Feature dimension:', ds.feature_dim)
if len(ds) > 0:
    x, y, sid = ds[0]
    print('One sample id:', sid)
    print('x shape:', x.shape)  # [3, feature_dim]
    print('label:', y.item())
```

```python
# Train using code from train_vit.py
# Keep it simple first; tune later if needed.
train(
    features_dir=FEATURES_DIR,
    artifacts_dir=ARTIFACTS_DIR,
    batch_size=16,
    num_workers=2,
)
```

```python
# Evaluate using code from evaluate_vit.py
evaluate(
    features_dir=FEATURES_DIR,
    artifacts_dir=ARTIFACTS_DIR,
    batch_size=16,
    num_workers=2,
)
```

```python
# Show output files
import glob

print('Checkpoints:')
for p in sorted(glob.glob(os.path.join(ARTIFACTS_DIR, 'checkpoints', '*'))):
    print(' -', p)

print('Metrics:')
for p in sorted(glob.glob(os.path.join(ARTIFACTS_DIR, 'metrics', '*'))):
    print(' -', p)
```
